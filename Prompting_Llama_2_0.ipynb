###1.2 Loading Model & Tokenizer 🧠

Here, we're loading both the Llama model and its associated tokenizer.
The tokenizer will assist in converting our text prompts into a format that the model can understand and process. 📝


```python
from transformers import AutoTokenizer
import transformers
import torch

model = "meta-llama/Llama-2-7b-chat-hf" # meta-llama/Llama-2-7b-hf

tokenizer = AutoTokenizer.from_pretrained(model, use_auth_token=True)
```

###1.3  Establishing the Llama Pipeline 🛠️

Let's set up a pipeline for text generation. 🚀 This pipeline simplifies the process of feeding prompts to our model and receiving generated text as output.



```python
from transformers import pipeline

llama_pipeline = pipeline(
    "text-generation",  # LLM task
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

###1.4  Engagement with Llama

Now that everything is set up, let's see how 🦙 Llama responds to some sample queries. 🎉


```python
def get_llama_response(prompt, do_sample=True, temperature=0.1, top_k=10, num_return_sequences=1, eos_token_id=tokenizer.eos_token_id, max_length=1024):
    """
    Generate a response from the Llama model.

    Parameters:
        prompt (str): The user's input/question for the model.
        do_sample (bool): Whether to use sampling during generation (default is True).
        temperature (float): Sampling temperature (default is 0.5).
        top_k (int): Number of top-k tokens to consider during sampling (default is 10).
        num_return_sequences (int): Number of response sequences to generate (default is 1).
        eos_token_id (int): ID of the end-of-sequence token (default is tokenizer.eos_token_id).
        max_length (int): Maximum length of the generated response (default is 1024).

    Returns:
        None: Prints the model's response.
    """
    # Uncomment the following lines to use the Llama model
    prompt = f"<s>[INST] {prompt} [/INST]"
    response = llama_pipeline(
         prompt,
         do_sample=do_sample,
         top_k=top_k,
         num_return_sequences=num_return_sequences,
         eos_token_id=eos_token_id,
         max_length=max_length,
         temperature=temperature,
     )
    return response


# Example usage:
prompt = 'I tell me how to play cricket ?\n'
respopnse = get_llama_response(prompt=prompt)
print("Chatbot:", respopnse[0]['generated_text'])
```

#2. Changing the temperature setting 🌡️

   - The code specifies three different temperature settings: 0.1, 0.5, and 1.0.
   - Temperature affects the randomness and creativity of the generated responses. Lower values (closer to 0) produce more deterministic and conservative responses, while higher values (closer to 1) introduce more randomness and diversity.


This approach provides insights into how temperature influences the generation of responses, helping to understand the behavior of the language model under different conditions.


```python
import time



def print_llama_responses(prompt, temperatures):
    for temperature in temperatures:
        print(f"----------------- Lalam 2.0 response with temperature = {temperature} -------------------")
        # Pause the code for 30 seconds
        response =get_llama_response(prompt=prompt, temperature=temperature)
        print("Chatbot:", response[0]['generated_text'])


# Define the prompt and list of temperatures
prompt = """Write a funncy coverl letter of 5 lines for me? It is about acting"""
temperatures = [0.1, 0.5, 1.0]

# Call the function to print responses for different temperatures
print_llama_responses(prompt, temperatures)
```

#3.  Changing the No of Token setting `max-length` 🔢

  - The `max_length` parameter controls the maximum length of the token sequence that Llama 2.0 generates in its response.
  - The code tests different values for `max_length` to observe how the length restriction impacts the generated responses.


Understanding how Llama 2.0 performs under various token length settings is crucial for optimizing its usage and leveraging its capabilities effectively.


```python
# Example usage:
prompt = "Tell me an interesting fact."
max_lengths_to_test = [50, 100, 200]

for max_len in max_lengths_to_test:
    respopnse =get_llama_response(prompt=prompt, max_length=max_len)
    print("Chatbot:", respopnse[0]["generated_text"])
    print("-" * 40)
```

### Consequences of Exceeding the max_length Parameter

When exceeding the `max_length` parameter, errors can occur, particularly if the input surpasses the maximum number of tokens supported by the model.

In the given scenario, attempting to summarize "The Velveteen Rabbit" text, which contains 7495 words, with a model that expects a maximum of 1024 tokens, may result in an error. ❌

To mitigate this issue, one can either split the input text into smaller segments or opt for a model with a higher token limit, if available. Segmenting the input enables processing in smaller chunks that fit within the model's token limit. Alternatively, utilizing a model with a higher token limit may allow processing of longer inputs without encountering errors. 🛠️

#4. Multi-turn chat in Llama 🔄


LLMs (Large Language Models) are stateless and do not remember the previous chat. They do not keep any record of it. 🧠❌. Let us check it with simple example.



```python
prompt = """
    I love mathematics? tell me about main areas in it.
"""
response =get_llama_response(prompt)
print("Chatbot:", response[0]["generated_text"])

prompt_2 = """
Which math area mentioned previously is more relevant to machine learning?
"""
response = get_llama_response(prompt_2)
print("Chatbot:", response[0]["generated_text"])
```

**After running the below code snippet we realize that 2nd response does not use anything from the first response. If we want then we have to explicity pas s the last response in the prompt**

```python
prompt_1 = """
    I love mathematics? tell me about main areas in it?
"""
response_1 = get_llama_response(prompt_1)
prompt_2 = """
Which math area mentioned previously is more relevant to machine learning?
"""
chat_prompt = f"""
<s>[INST] {prompt_1} [/INST]
{response_1[0]['generated_text']}
</s>
<s>[INST] {prompt_2} [/INST]
"""
print(chat_prompt)


```

```python
response_2 = get_llama_response(chat_prompt)
print("Chatbot:", response_2[0]["generated_text"])
```

#5. Zero-shot Prompting🎯
- Here is an example of zero-shot prompting.
- In this scenario, you’re providing a prompt to the model without explicitly giving any examples of the completed task.
- The goal is to test whether the model can infer the task based solely on the structure of your prompt.
- Zero-shot prompting relies on providing the prompt’s structure, but without any pre-existing task instances.

```python
prompt = """
Message: Hi Abdul, I appreciate the lunch, it was delicious!
Sentiment: ?
"""
respopnse = get_llama_response(prompt=prompt)
print("Chatbot:", respopnse[0]['generated_text'])
```

#6.  Few-shot Prompting🎯
- Here is an example of few-shot prompting.
- In the context of few-shot prompting, you not only outline the structure of the task in your prompt but also include two or more specific examples.
- The goal is to assess whether the model can deduce the task based on both the prompt’s structure and the provided examples.
- Few-shot prompting leverages a combination of structure and exemplars to guide the model’s understanding.

```python
prompt = """
Message: Hi ! How is life going on!
Sentiment: Neutral

Message: I love studying for machcine learning
Sentiment: Positive

Message: I hate pumkin pie!
Sentiment: ?
"""
respopnse = get_llama_response(prompt=prompt)
print("Chatbot:", respopnse[0]['generated_text'])
```

```python
prompt = """
Message: Hi ! How is life going on!
Sentiment: Neutral

Message: I love studying for machcine learning
Sentiment: Positive

Message: I hate pumkin pie!
Sentiment: ?

Give a one word response, either Postivie Negative or Neutral.
"""
respopnse = get_llama_response(prompt=prompt)
print("Chatbot:", respopnse[0]['generated_text'])
```

#7. Role Prompting 🦸
