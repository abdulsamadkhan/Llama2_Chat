### 3️⃣ Loading Model & Tokenizer 🧠

Here, we're loading both the Llama model and its associated tokenizer.
The tokenizer will assist in converting our text prompts into a format that the model can understand and process. 📝


```python
from transformers import AutoTokenizer
import transformers
import torch

model = "meta-llama/Llama-2-7b-chat-hf" # meta-llama/Llama-2-7b-hf

tokenizer = AutoTokenizer.from_pretrained(model, use_auth_token=True)
```

### 4️⃣ Establishing the Llama Pipeline 🛠️

Let's set up a pipeline for text generation. 🚀 This pipeline simplifies the process of feeding prompts to our model and receiving generated text as output.

### ❗ `torch.d_type` parameter
- `torch.float32` or `torch.float`: Default in PyTorch, balances precision and speed. 🏃‍♂️
- `torch.float16` or `torch.half`: Uses less memory and resources, but less precise. Can speed up models on specific hardware. 🚀
- `torch.float64` or `torch.double`: More precise but requires more resources. 🎯

Note: Not all models and operations support all data types. Changing data types requires careful testing and checking the documentation. 📚

There's no `torch.float8` in PyTorch. For lower precision, consider integer types `torch.int8` or `torch.uint8`, but beware of precision loss and limited support. ⚠️



```python
from transformers import pipeline

llama_pipeline = pipeline(
    "text-generation",  # LLM task
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

###  Engagement with Llama

Now that everything is set up, let's see how 🦙 Llama responds to some sample queries. 🎉


```python
def get_llama_response(prompt: str) -> None:
    """
    Generate a response from the Llama model.

    Parameters:
        prompt (str): The user's input/question for the model.

    Returns:
        None: Prints the model's response.
    """
    sequences = llama_pipeline(
        prompt,
        do_sample=True,
        top_k=10,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        max_length=1024,
    )
    print("Chatbot:", sequences[0]['generated_text'])


prompt = 'Can you tell me how to create a pop function in list ?\n'
get_llama_response(prompt)
```

# do not repeat my question ?
If you'd rather not repeat your question to the llama and only receive its response, you can achieve this by setting return_full_text to False. 🦙

```python
def get_llama_response(prompt: str) -> None:
    """
    Generate a response from the Llama model.

    Parameters:
        prompt (str): The user's input/question for the model.

    Returns:
        None: Prints the model's response.
    """
    sequences = llama_pipeline(
        prompt,
        do_sample=True,
        top_k=10,
        num_return_sequences=1,
        return_full_text=False, # to not repeat the question, set to False
        eos_token_id=tokenizer.eos_token_id,
        max_length=1024,
    )
    print("Chatbot:", sequences[0]['generated_text'])


prompt = 'Can you tell me how to create a pop function in list ?\n'
get_llama_response(prompt)
```

# Other important parameters for the Llama pipeline
1. `repetition_penalty`

The repetition_penalty is a parameter used in the generation process of language models like Llama. It’s designed to control the model’s tendency to repeat itself.

When the repetition_penalty is set to 1.0, it means there’s no penalty for repetition, and the model is free to repeat the same phrases or sentences multiple times if it determines that to be the most likely sequence.

If you increase the repetition_penalty to a value greater than 1, the model becomes less likely to repeat itself. This is because the penalty multiplies the probabilities of tokens that have already appeared, making them less likely to be chosen again.

2. `top_k`

parameter is used in the generation process of language models for what’s known as top-k sampling.

For example, if top_k is set to 50, at each step, the model will only consider the 50 tokens that it thinks are most likely to come next. One of these 50 tokens is then randomly chosen as the next token, with the probability of each token being chosen corresponding to its predicted likelihood.

Remember, these settings can greatly affect the quality and usefulness of the generated text, so it’s important to experiment and find the best values for your specific use case.

3. `do_sample`:

This is a boolean parameter that determines whether sampling is used for generating text. If do_sample is set to True, the model will use probabilistic sampling to generate each token, which introduces randomness and can lead to more diverse and creative output.

If do_sample is set to False, the model uses greedy decoding, which always selects the most probable next token. Greedy decoding can make the output more deterministic and coherent, but it may also be more repetitive and less creative.
4. `temperature`:

This is a float parameter that controls the randomness of the predictions. A higher temperature value (e.g., closer to 1) makes the output more random, while a lower temperature value (e.g., closer to 0) makes the output more deterministic. Specifically, the temperature parameter adjusts the probabilities of the next tokens: a high temperature flattens the probability distribution and makes unlikely tokens more likely, while a low temperature sharpens the distribution and makes the most likely tokens even more likely.

5. `max_new_tokens`

parameter is used in the generation process of language models. It specifies the maximum number of tokens that can be generated in a single sequence.

For example, if max_new_tokens is set to 500, the model will stop generating more tokens once it has generated 500 tokens, even if it hasn’t reached an end-of-sequence token. This can be useful for controlling the length of the generated text and ensuring that it doesn’t become excessively long.

6. `max_length`

 — The maximum length the generated tokens can have. Corresponds to the length of the input prompt + max_new_tokens. Its effect is overridden by max_new_tokens, if also set.
